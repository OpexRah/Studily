```python
# API client library
import googleapiclient.discovery
import re
# API information
api_service_name = "youtube"
api_version = "v3"
# API key

DEVELOPER_KEY = "ENTER KEY"
# API client
youtube = googleapiclient.discovery.build(api_service_name, api_version, developerKey = DEVELOPER_KEY)
# 'request' variable is the only thing you must change
# depending on the resource and method you need to use
keyword_list = ["Pranks","memes","NBA","parties","anime","mrbeast"]
titles = []
ids = []
# in your query
for keyword in keyword_list:
  request = youtube.search().list(
          part="id,snippet",
          type='video',
          fields="items(id(videoId),snippet(title,description))",
          q=keyword,
          maxResults=60
  )
  # Query execution
  response = request.execute()
  # Print the results
  #print(response)

  for i in range(0,50):
    final_title = ""
    title = response["items"][i]["snippet"]["title"]
    title_word_list = title.split(" ")
    for word in title_word_list:
      #word_filtered = ''.join(filter(str.isalnum, word))
      pattern = r'[^A-Za-z0-9]+'
      word_filtered = re.sub(pattern, '', word)
      final_title += word_filtered + " "
    id = response["items"][i]["id"]["videoId"]
    titles.append(final_title)
    ids.append(id)
    print(final_title)
```

```python
import csv
data_file = open('./jsonoutput.csv', 'w', newline='')
items = response["items"]
csv_writer = csv.writer(data_file)
csv_writer.writerow(["id", "title"])

for i in range(0, len(titles)):
  new = [ids[i], titles[i]]
  csv_writer.writerow(new)

data_file.close()
```
